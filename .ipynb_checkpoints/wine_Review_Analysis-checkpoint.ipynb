# Analysis

### generally there seems to be a positive correlation between the average price of a bottle of wine and the points it receives from Wine Enthusiast. The linear regression seems to be a poor fit for this data. The R-squared of .37 supports this. The logarithmic regression would be a better fit from the shape of the data.



```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as stats
import statsmodels.api as sm
from scipy.stats import linregress
```

```python
# Import CSV File and show the first 5 rows
wine_data_csv = "CSV/winemag_data_pandata.csv"
wine_data = pd.read_csv(wine_data_csv)
wine_data.head()
```

```python
# determine how many initial records there are
len(wine_data)
```

```python
# Count the reviews by country
country_count = wine_data.groupby(["country"]).count()["designation"]
country_count = pd.DataFrame(country_count)
country_count = country_count.reset_index()
len(country_count)
```

```python
# Select only instances where a country has more than 1000 reviews
country_count = country_count[country_count['designation'] > 1000]
country_count = country_count.sort_values('designation',ascending=False)
country_count

```

```python
# Create a dataframe that groups by country and price point and takes the avg points scored
wine_price = wine_data.groupby(["country", "price"]).mean()["points"]
wine_price = pd.DataFrame(wine_price)
wine_price = wine_price.reset_index()
wine_price.head(10)
```

```python
# Resetting the index of the df so I can use loc
wine_price_limit  = wine_price.set_index('country')
wine_price_limit
```

```python
# Narrowing the final dataset to only countries with greater than 1000 reviews
wine_price_limit = wine_price_limit.loc[['US','France', 'Italy', 'Spain', 'Portugal', 'Chile', 'Argentina', 'Austria', 'Germany', 'Australia']]
wine_price_limit

```

```python
# Removing any instances of negative prices
wine_price_limit = wine_price_limit[wine_price_limit['price'] > 0]
wine_price_limit
```

```python
# Plot the relationship between the avg price vs the points using a scatterplot
sns.scatterplot(data=wine_price_limit, x='price',y='points',color='#73cae7', label='price vs. points')

plt.xlabel("AVG Price")
plt.ylabel("AVG Points")
plt.legend=('Best')
plt.gcf().set_size_inches(14,7)
```

```python
# Remove outliers from the data where price is greater than $1500
wine_price_limit = wine_price_limit[wine_price_limit['price'] < 1500]
wine_price_limit
```

```python
# Plot the relationship between the avg price vs the points using a scatterplot
sns.scatterplot(data=wine_price_limit, x='price',y='points',  color='#73cae7', label='price vs. points')



plt.xlabel("AVG Price")
plt.ylabel("AVG Points")
plt.legend=('Best')
plt.gcf().set_size_inches(14,7)
```

```python
price = wine_price_limit['price']
points = wine_price_limit['points']
linregress(price,points) 
```

```python
# Add linear regression line to the graph
(slope, intercept, _, _, _) = linregress(price, points)
fit = slope * price + intercept

sns.scatterplot(data=wine_price_limit, x='price',y='points',  color='#73cae7', label='price vs. points')
sns.lineplot(price, fit)


plt.xlabel("AVG Price")
plt.ylabel("AVG Points")
plt.legend=('Best')
plt.gcf().set_size_inches(14,7)
```

```python
#fit model for OLS statsmodels regression

wine_price_limit = wine_price_limit[wine_price_limit.replace([np.inf, -np.inf], np.nan).notnull().all(axis=1)]

X = wine_price_limit['price']
y = wine_price_limit["points"]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

```python
# Add polynomial regression line to the graph
(slope, intercept, _, _, _) = linregress(price, points)
fit = np.polyfit(price, points, deg=1)

sns.scatterplot(data=wine_price_limit, x='price',y='points',  color='#73cae7', label='price vs. points')
sns.lmplot(data=wine_price_limit, x='price', y='points', lowess=True, ci=None, scatter_kws={"s": 80}, line_kws={'color': 'red'});



plt.xlabel("AVG Price")
plt.ylabel("AVG Points")
plt.legend=('Best')
plt.gcf().set_size_inches(14,7)
```
